#EXPLORATORY DATA ANALYSIS

```python
# Loading and Reading the Dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content.drive')

df = pd.read_csv("/content.drive/MyDrive/iris.csv")
df
```

###DATA VISUALIZATION
1. Histogram

```python
#Visualize the data using a histogram, A single variable and the variable count
df.hist()
plt.show()
```

2. Scatter Plots

```python
#Relationship between petal length and species
plt.scatter(df['petal_length'], df['species'])
plt.xlabel('petal_length')
plt.ylabel('species')
plt.show()
```

```python
#Relationship between petal width and species
plt.scatter(df['petal_width'], df['species'])
plt.xlabel('petal_length')
plt.ylabel('species')
plt.show()
```

```python
#Relationship between sepal length and species
plt.scatter(df['sepal_length'], df['species'])
plt.xlabel('sepal_length')
plt.ylabel('species')
plt.show()
```

```python
#Relationship between sepal width and species
plt.scatter(df['sepal_width'], df['species'])
plt.xlabel('sepal_width')
plt.ylabel('species')
plt.show()
```

```python
#Relationship between petal length and petal width
plt.scatter(df['petal_length'], df['petal_width'])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

3. Boxplot

```python
#Relationship between petal length and the species
sns.boxplot(x='species',y='petal_length',data=df)
plt.show()
```

```python
#Relationship between petal width and species
sns.boxplot(x='species',y='petal_width',data=df)
plt.show()
```

```python
#Relationship between sepal length and species, boxplot
sns.boxplot(x='species',y='sepal_length',data=df)
plt.show()
```

```python
#Relationship between sepal width and species
sns.boxplot(x='species',y='sepal_width',data=df)
plt.show()
```

#Identifying correlations between numerical features

```python
#Correlation Heatmap
plt.figure(figsize=(10, 6))
numeric_df = df[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt= '.2f')
plt.title('Correlation Heatmap of Numeric Variables')
plt.tight_layout()
plt.show()
```
